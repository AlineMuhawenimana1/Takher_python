### Exercise 44:
Create a waterfall chart using Matplotlib to show the changes in a sequence of values.


```python
# Exercise 44
import matplotlib.pyplot as plt
import numpy as np

data = [1000, 300, -150, -200]  
labels = ['Start', 'Sales', 'Returns', 'Costs', 'End']
steps = np.insert(np.cumsum(data), 0, 0) 

bars = np.diff(steps)
starts = steps[:-1]

colors = ['gray'] + ['green' if v >= 0 else 'red' for v in bars[1:]] + ['gray']
bars = np.append(bars, steps[-1])  
starts = np.append(starts, 0)      

plt.bar(labels, bars, bottom=starts, color=colors)
plt.title('Waterfall Chart')
plt.show()
```
